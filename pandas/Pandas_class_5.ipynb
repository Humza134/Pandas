```python
data.kms_driven.replace("Petrol", 0, inplace=True)
# data.kms_driven.apply(lambda x: 0 if x == "Petrol" else x)
```

```python
data.kms_driven.unique()
```

```python
data.kms_driven = data.kms_driven.str.replace(",", "").str.replace(" kms", "").fillna(0).astype(int)
```

```python
data.kms_driven.isna().sum()
```

```python
data.kms_driven = data.kms_driven.replace(0,data.kms_driven.median())
```

```python
data.Price.unique()
```

```python
data.Price = data.Price.replace("Ask For Price", "0").str.replace(",", "").astype(int)
```

```python
data.Price.replace(0, data.Price.median(), inplace=True)
```

```python
data.fuel_type.unique()
```

```python
data.fuel_type.isna().sum()
```

```python
data.fuel_type = data.fuel_type.fillna(data.fuel_type.mode()[0])
```

```python
data.year.unique()
```

```python
data.year.value_counts()
```

```python
import numpy as np
def transform_year(year):
    if year.isalpha():
        return np.nan
    else:
        if year.isnumeric():
            return int(year)
```

```python
data.year = data.year.apply(transform_year)
```

```python
data.year = data.year.fillna(method='ffill')
```
