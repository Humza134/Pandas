# Data Wrangling: Join, Combine , Reshape

```python
data0 = pd.read_excel("/content/test_data.xlsx", sheet_name=0)
data1 = pd.read_excel("/content/test_data.xlsx", sheet_name=1)
data2 = pd.read_excel("/content/test_data.xlsx", sheet_name=2)
```

```python
data0
```

```python
data1
```

```python
data2
```

```python
pd.concat([data0, data1]).reset_index(drop=True)
```

```python
pd.concat([data0, data1], axis=1)
```

```python
pd.merge(data0,data1, how = "inner", on= "sales_man_id")
```

```python
pd.merge(data0,data1, how = "inner", on= "sales_man_name")
```

```python
pd.merge(data1, data2, on="sales_man_name", how="inner")
```

```python
pd.merge(data1, data2, on="sales_man_name", how="outer")
```

```python
pd.merge(data2, data1, how='left')
```

```python
pd.merge(data2, data1, how='right')
```

#Pivot

```python
df = pd.DataFrame({'NAME': ['John', 'Boby', 'Mina', 'Peter', 'Nicky'],
      'DEGREE': ['Masters', 'Graduate', 'Graduate', 'Masters', 'Graduate'],
      'AGE': [27, 23, 21, 23, 24]})
```

```python
df
```

```python
table = pd.pivot_table(df, index =['NAME', 'DEGREE'])
table
```

```python
# values can be an object or a list
df.pivot (index="NAME",columns = "DEGREE", values = "AGE")
```
