```python
# Calculate the mean of the 'Deaths' column, ignoring non-numeric values
mean_deaths = pd.to_numeric(data['Deaths'], errors='coerce').mean()

# Replace non-numeric values in the 'Deaths' column with the mean
data['Deaths'] = pd.to_numeric(data['Deaths'], errors='coerce').fillna(mean_deaths)
```
